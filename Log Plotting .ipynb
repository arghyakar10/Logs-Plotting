```python
# plotted SP log vs Depth

plt.style.use("default")
plt.figure(figsize=(4,8))
plt.plot(df['SP'],df['Depth'])
plt.ylim(5000,0)
plt.grid()
plt.axvline(x=-100,color='red')
```

```python
#plotted deep and low resistivity vs depth

plt.style.use('default')
plt.figure(figsize=(4,8))
plt.plot(np.log10(df['RILD']),df['Depth'],label='Deep Resistivity')
plt.plot(np.log10(df['RLL3']),df['Depth'],label='Low Resistivity')
plt.legend(loc="upper right")
plt.grid()
plt.ylim(5000,0)
plt.axhspan(1500,2500,color='black')
```

```python
plt.figure(figsize=(4,8))
plt.style.use('default')
plt.plot(df['SP'],df['Depth'],label='SP')
plt.plot(df['GR'],df['Depth'],label='GR')
plt.axhspan(3950,4642,color='black',label='Shale Bed')
plt.grid()
plt.ylim(5000,0)
plt.legend(loc='upper right')
```

```python
grmax = max(df['GR'])
grmin = min(df['GR'])

# Applied vol. of shale formula:
df['Vsh'] = (df['GR'] - grmin)/(grmax - grmin) 

# Cosmetic changes
plt.style.use('default')
plt.figure(figsize=(4,8))

# Here plotted the both graphs:
plt.plot(df['GR'],df['Depth'],label='GR')
plt.plot(df['Vsh']*100,df['Depth'],label='Vol. of Shale')

plt.ylim(5000,0)
plt.legend(loc='best')
plt.grid()
```

```python
# Cosmetic changes
plt.style.use('default')
plt.figure(figsize=(4,16))

# Here plotted the both graphs:
plt.plot(df['DPOR'],df['Depth'],label='Density Porosity')
plt.plot(df['CNLS'],df['Depth'],label='Neutron Porosity')

plt.ylim(5000,0)
plt.legend(loc='best')
plt.grid()
```

```python
plt.style.use('default')
plt.figure(figsize=(4,16))

# Here plotted the both graphs:
plt.plot(df['RHOB'],df['Depth'],label='Density Porosity')


plt.ylim(5000,0)
plt.legend(loc='best')
plt.grid()
```

```python
plt.style.use('default')
plt.figure(figsize=(4,16))

# Here plotted the both graphs:
plt.plot(np.log(df['GR']),df['Depth'],label='Density Porosity')
plt.plot(df['DCAL'],df['Depth'],label='Caliper Log')

plt.ylim(5000,0)
plt.legend(loc='best')
plt.grid()
```

```python
plt.figure(figsize=(4,16))
fig, ax = plt.subplots(nrows= 1, ncols = 4,sharey=True)

fig.set_figheight(15)
fig.set_figwidth(15)

# Here Axes starts from 0
# Each axis plotting describes different logs:

ax[0].plot(df['DCAL'],df['Depth'],label='Caliper')
ax[0].set_ylim(5000,0)
ax[0].set_title('Caliper Log')
ax[0].axhspan(3500,5000,color='red',alpha=0.3,label='Wash-Out')
ax[0].legend()
ax[0].grid()

ax[1].plot(df['SP'],df['Depth'],label='SP')
ax[1].plot(df['GR'],df['Depth'],label='GR')
ax[1].legend(loc='upper center')
ax[1].set_title('Lithology Logs')
ax[1].axhspan(3500,5000,color='gold',alpha=0.3)
ax[1].grid()

ax[2].plot(np.log(df['RILD']),df['Depth'],label='RILD')
ax[2].plot(np.log(df['RLL3']),df['Depth'],label='RLL3')
ax[2].legend(loc='upper center')
ax[2].set_title('Resistivity Logs')
ax[2].axhspan(3500,5000,color='gold',alpha=0.3)
ax[2].axhspan(1500,2500,color='green',alpha=0.3)
ax[2].grid()

ax[3].plot(df['DPOR'],df['Depth'],label='DPOR',color='red')
ax[3].plot(df['CNLS'],df['Depth'],label='CNLS',color='purple')
ax[3].legend(loc='upper center')
ax[3].set_title('Porosity Logs')
ax[3].axhspan(3500,5000,color='black',alpha=0.1)
ax[3].axhspan(1500,2500,color='green',alpha=0.3)
ax[3].axhspan(2500,3500,color='orange',alpha=0.6)
ax[3].grid()
```
